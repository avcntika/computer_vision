```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (20,10)
plt.rcParams['font.size'] = 14
```

```python
img = cv2.imread('images/pano1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
height, width = img.shape[:2]
plt.imshow(img)
plt.show()
```

OpenCV Affine Transformations

Translation

```python
M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img, M, (width,height))
```

```python
plt.imshow(dst)
```

Rotation about origin

```python
angle = np.radians(20)
R = np.array([
    [np.cos(angle), np.sin(angle), 0],
    [-np.sin(angle), np.cos(angle), 0]
])
dst = cv2.warpAffine(img, R, (width,height))
```

```python
plt.imshow(dst)
```

Shearing

```python
M = np.float32([[1,0.5,0],[0,1,0]])
dst = cv2.warpAffine(img, M, (width,height))
```

```python
plt.imshow(dst)
```

Rotation about centre

```python
R = cv2.getRotationMatrix2D(center=(width//2, height//2), angle=20, scale=1)
dst = cv2.warpAffine(img, R, (width,height))
```

```python
plt.imshow(dst)
```

Scaling

```python
scale = 10
S = np.array([
    [scale, 0, 0],
    [0, scale, 0]
],dtype=np.float32)
dst = cv2.warpAffine(img, S, (width*scale,height*scale))
```

```python
plt.imshow(dst)
```
